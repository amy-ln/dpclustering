```python
import pandas as pd
from lloyd import lloyd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate synthetic data
data1 = []
for n in range(100):
    data1.append(np.random.multivariate_normal(mean=[2,8], cov=[[1,0],[0,1]]))
df1 = pd.DataFrame(data1, columns=["x","y"])
data2 = []
for n in range(100):
    data2.append(np.random.multivariate_normal(mean=[3,4], cov=[[1,0],[0,1]]))
df2 = pd.DataFrame(data2, columns=["x","y"])
```

```python
df = pd.concat([df1,df2])
plt.scatter(df["x"], df["y"])
```

```python
centers = lloyd(2, df, 10)
centers_df = pd.DataFrame(centers, columns = ["x", "y"])
plt.scatter(df["x"], df["y"])
plt.scatter(centers_df["x"], centers_df["y"], color = "red")
plt.show()
```
